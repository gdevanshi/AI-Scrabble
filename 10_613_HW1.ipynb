```python
# Load the German dataset
from aif360.datasets import GermanDataset
dataset = GermanDataset(protected_attribute_names=["sex"],
privileged_classes=[lambda x: x == "male"],
features_to_drop=["personal_status", "age"]
)
# Split the dataset into training and validation sets
dataset_train, dataset_val = dataset.split([0.8], shuffle=True)
#Get features and labels
X_train,y_train = dataset_train.features, dataset_train.labels.ravel()
X_val,y_val = dataset_val.features, dataset_val.labels.ravel()
```

```python
# Problem 1.1 Getting Familiar with the Dataset
print(len(dataset_train.feature_names))
print(dataset_train.feature_names)
print(dataset_train.label_names)
print(dataset_train.protected_attribute_names)
df = dataset.convert_to_dataframe()[0]
#print(df.loc[:,:])
print('month', df.month.unique())
#print('credit_amount', df.credit_amount.unique())
print('investment_as_income_percentage', df.investment_as_income_percentage.unique())
print('residence_since', df.residence_since.unique())
print('number_of_credits', df.number_of_credits.unique())
print('people_liable_for', df.people_liable_for.unique())
print('status=A12', df.loc[:, 'status=A12'].unique())
print("males", len(df[df.sex == 1]))
print("females", len(df[df.sex == 0]))
print("total", len(df))
print("range of credit amount", df.credit_amount.min(), df.credit_amount.max(), df.credit_amount.mean())
print("range of employment", len(df[df.loc[:, 'employment=A71'] == 1]), len(df[df.loc[:, 'employment=A72'] == 1]), len(df[df.loc[:, 'employment=A73'] == 1]), len(df[df.loc[:, 'employment=A74'] == 1]), len(df[df.loc[:, 'employment=A75'] == 1]))
print("range of checking acct", len(df[df.loc[:, 'status=A11'] == 1]), len(df[df.loc[:, 'status=A12'] == 1]), len(df[df.loc[:, 'status=A13'] == 1]), len(df[df.loc[:, 'status=A14'] == 1]))
print("foreign workers", len(df[df.loc[:, 'foreign_worker=A201'] == 1]), len(df[df.loc[:, 'foreign_worker=A202'] == 1]))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics 
from sklearn.model_selection import GridSearchCV
hyperparams = {
    'n_estimators': [50, 100, 150, 200],
    'max_features': ['auto', 'sqrt', 'log2', None],
    'max_depth': [5, 7, 9, 10, 12],
    'max_leaf_nodes': [3, 5, 9, 10, 12, 15],
}
classifier = RandomForestClassifier()
grid_search = GridSearchCV(estimator=classifier, param_grid=hyperparams, scoring='accuracy')
model = grid_search.fit(X_train, y_train)
grid_search.best_params_
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics 
classifier = RandomForestClassifier(max_depth=12, max_leaf_nodes=15, n_estimators=200, max_features=None)
model = classifier.fit(X_train, y_train)
pred = model.predict(X_val)
print(metrics.classification_report(y_val, pred))
pred_train = model.predict(X_train)
train_accuracy = metrics.accuracy_score(y_train, pred_train)
print("Training ccuracy:", train_accuracy)
#Overall val accuracy
val_accuracy = metrics.accuracy_score(y_val, pred)
print("Validation ccuracy:", val_accuracy)
```

```python
# Create a copy of the validation set with your classifier's predicted labels
dataset_val_pred = dataset_val.copy()
privileged_classes = [{"sex": 1}] # Men (sex = 1)
unprivileged_classes = [{"sex": 0}] # Women (sex = 0)
dataset_val_pred.labels = pred
```

```python
from aif360.metrics import ClassificationMetric
disparate_impact = ClassificationMetric(dataset_val, dataset_val_pred, unprivileged_groups = unprivileged_classes, privileged_groups = privileged_classes).disparate_impact()
print("Disparate Impact: ", disparate_impact)
equal_opportunity_diff = ClassificationMetric(dataset_val, dataset_val_pred, unprivileged_groups = unprivileged_classes, privileged_groups = privileged_classes).equal_opportunity_difference()
print("Equal Opportunity Difference: ", equal_opportunity_diff)
stat_parity_diff = ClassificationMetric(dataset_val, dataset_val_pred, unprivileged_groups = unprivileged_classes, privileged_groups = privileged_classes).statistical_parity_difference()
print("Statistical Parity Difference: ", stat_parity_diff)
ave_odds_diff = ClassificationMetric(dataset_val, dataset_val_pred, unprivileged_groups = unprivileged_classes, privileged_groups = privileged_classes).average_odds_difference()
print("Average Odds Difference: ", ave_odds_diff)
theil_index = ClassificationMetric(dataset_val, dataset_val_pred, unprivileged_groups = unprivileged_classes, privileged_groups = privileged_classes).theil_index()
print("Theil Index: ", theil_index)
```
